```python
Understanding Purchase Numbers Distribution. let's look into the distribution of our key metric, the 'Purchase Number.' We aim to understand how it's spread out. Your task is to conduct a Normality check on this metric for both groups and jot down the results. Let's study.
```

```python
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Perform Shapiro-Wilk test for normality
shapiro_control = shapiro(control_df['Purchase Number'])

shapiro_test = shapiro(test_df['Purchase Number'])

print("Shapiro-Wilk Test for Control Group:", shapiro_control)
print("Shapiro-Wilk Test for Test Group:", shapiro_test)

# Visualize the distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(control_df['Purchase Number'], kde=True)
plt.title('Control Group Purchase Number Distribution')

plt.subplot(1, 2, 2)
sns.histplot(test_df['Purchase Number'], kde=True)
plt.title('Test Group Purchase Number Distribution')

plt.show()
```

'll use the Shapiro-Wilk test and also visualize the distributions with histograms to make it easier to understand.

Great! Now that we've checked for normality, we can see from the Shapiro-Wilk test that the p-value for the control group is greater than 0.05, which means we can assume the data is normally distributed. However, the p-value for the test group is less than 0.05, indicating that it's not normally distributed.

Given this, the next step should be to use a non-parametric test to compare the "Purchase Number" between the two groups, since one of them is not normally distributed

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(control_df['Purchase Number'], test_df['Purchase Number'])

print("Mann-Whitney U test statistic:", statistic)
print("P-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("\nConclusion: The new design resulted in a statistically significant change in the number of purchases.")
else:
    print("\nConclusion: There is no statistically significant difference in the number of purchases between the two designs.")
```

Interpretation:  Let's break down how to interpret the signals from the Mann-Whitney U test we just ran:

1. The Mann-Whitney U test statistic (455.0):

This number itself isn't directly interpretable in a business context. It's a calculated value that represents the difference in the ranking of the "Purchase Number" between the two groups. A larger value would indicate a bigger difference in the rankings, but it's not a percentage or a direct measure of effect. You can think of it as an intermediate calculation that the test uses to determine the p-value.
2. The P-value (0.94695...):

This is the most important number for decision-making. The p-value tells us the probability of seeing the observed difference in purchases between the two groups (or an even more extreme difference) if there was actually no real difference between them.
In our case, the p-value is very high (approximately 0.95 or 95%). This means that there is a 95% chance of seeing this kind of result purely by random chance, even if the new design has no real effect on purchases.
The 0.05 Rule of Thumb: In business and research, we usually use a p-value of 0.05 (or 5%) as a cutoff. If the p-value is less than 0.05, we say the result is "statistically significant," meaning it's unlikely to be due to random chance. Since our p-value is much higher than 0.05, we conclude that the result is not statistically significant.
In a Nutshell for a Non-Technical Audience:

The high p-value tells us that the small difference we saw in the number of purchases between the old and new designs is very likely just random noise, not a real effect of the design change. Therefore, we can't be confident that the new design is any better or worse than the old one when it comes to driving purchases.

```python
from scipy.stats import ttest_ind

# Perform independent t-test
t_statistic, p_value = ttest_ind(control_df['Purchase Number'], test_df['Purchase Number'])

print("T-test statistic:", t_statistic)
print("P-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("\nConclusion (from t-test): The new design resulted in a statistically significant change in the number of purchases.")
else:
    print("\nConclusion (from t-test): There is no statistically significant difference in the number of purchases between the two designs.")
```

As you can see, the t-test also gives a very high p-value (approximately 0.976), which leads to the same conclusion as the Mann-Whitney U test: there is no statistically significant difference in the number of purchases between the two designs.

Interpretation of the t-test results:

T-test statistic (0.0307): This value is very close to zero, which indicates that the means of the two groups are very similar.
P-value (0.9756): This extremely high p-value reinforces that the small difference in the average number of purchases between the two groups is almost certainly due to random chance.
Moving Forward:

Based on these findings, we recommend **not to launch the new design universally at this time.** Instead, we should iterate on the new design to improve its conversion rate while maintaining its high level of engagement.

## 2. Introduction

The objective of this A/B test was to compare the performance of two watch screen designs:

*   **Control Group (Design A):** The existing watch screen design.
*   **Test Group (Design B):** A new design with a different call-to-action (CTA).

**Hypotheses:**

*   **Null Hypothesis (H0):** The new design does not yield a significant improvement in marketing campaign performance.
*   **Alternative Hypothesis (H1):** The new design outperforms the existing design.

---

## 3. Methodology

We analyzed data from two separate campaigns, one for each design, over a 30-day period. The following Key Performance Indicators (KPIs) were used to evaluate the performance of each design:

*   **Primary KPI:** Return on Investment (ROI)
*   **Secondary KPIs:**
    *   Click-Through Rate (CTR)
    *   Conversion Rate
    *   Cost per Conversion
    *   Cost Per Click (CPC)

---

## 4. Statistical Analysis

To ensure the validity of our conclusions, we employed the following statistical tests:

*   **Shapiro-Wilk Test:** We used this test to check for the normality of the data distribution for our key metrics. This is a crucial step in determining the appropriate hypothesis test to use.
*   **Independent t-test:** We used this test to compare the means of the key performance indicators (KPIs) between the control and test groups. The t-test is a robust statistical test for comparing the means of two independent groups, assuming that the data is normally distributed.
*   **Mann-Whitney U Test:** We also considered the Mann-Whitney U test, a non-parametric alternative to the t-test, for the metrics that did not meet the normality assumption. This ensures that our conclusions are not biased by the choice of statistical test.

By using a combination of these tests, we can be confident in the statistical significance of our findings.

## 6. Conclusion

Based on our primary KPI, Return on Investment (ROI), we **fail to reject the Null Hypothesis (H0)**. We did not find a statistically significant improvement in profitability with the new design.

While the new design showed a significant improvement in a secondary KPI (CTR), this engagement did not translate into a better overall business outcome (ROI).
