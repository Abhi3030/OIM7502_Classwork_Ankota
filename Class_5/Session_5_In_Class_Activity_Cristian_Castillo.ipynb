**5. Using Taxes as the independent variable and Sell as the dependent variable, fit a linear regression model and display the output from the model**

```python
X = sm.add_constant(df[' "Taxes"'])

y = df['Sell']

x = sm.add_constant(X)

model = sm.OLS(y, X).fit()

model.summary()
```

**6. In addition to the model output include markdown commentary about the important statistics from the output; is taxes a good predictor for selling price? Why or why not?**

The regression output provides insights into the relationship between the dependent variable 'Sell' (selling price) and the independent variable 'Taxes'.

<u>Coefficient Interpretation:</u>
On average, for each one-unit increase in 'Taxes', the selling price ('Sell') increases by 0.0414 units.

<u>Statistical Significance:</u>
The p-value associated with 'Taxes' is very low, indicating that the relationship between 'Taxes' and Selling Price is statistically significant.

<u>R-squared:</u>
The R-squared value is 0.809, suggesting that approximately 80.9% of the variability in Selling Price can be explained by the model.

<u>Conclusion:</u>
The regression model is statistically significant.'Taxes' appears to be a good predictor for Selling Price given its low p-value.

**7. Repeat this procedure using one of the other variables as the independent variable. Which model is "better"? What are some limitations of the model?** 

```python
X_beds = df[' "Beds"']
y = df['Sell']

X_beds = sm.add_constant(X_beds)


model_beds = sm.OLS(y, X_beds).fit()


model_beds.summary()
```

The regression with Taxes as the independent variable has a much higher R-squared value (0.809), indicating a better fit to the data compared to the model with Bedrooms (0.112).

The F-statistic for the Taxes model is significantly higher, suggesting a more significant overall relationship.

The p-value for 'Taxes' is very low, indicating that Taxes is a statistically significant predictor of 'Sell'.
On the other hand, the p-value for 'Beds' is also low, but not as low as 'Taxes'.

Conclusion:

The model with Taxes as the independent variable is better based on R-squared, F-statistic, and the significance of the predictor variable.
