### Function to Fetch API Data

```python
# Function to fetch data from a specific endpoint
def fetch_data_from_endpoint(endpoint):
    # request URL
    url = f"{base_url}/{endpoint}?apikey={api_key}"
    # GET request
    response = requests.get(url)
    if response.status_code == 200:
        # Getting JSON file and Converting to pandas DataFrame
        data = response.json()
        df = pd.DataFrame(data)
        return df
    else:
        print(f"Failed to fetch data: {response.status_code}, Response: {response.text}")
        return None
```

### Using Endpoints - Applicants

```python
# applicants usage
endpoint = 'applicants'  # You can change this to 'jobs', 'categories', etc.
df = fetch_data_from_endpoint(endpoint)

if df is not None:
    # Save the df to a CSV file
    df.to_csv('fetched_applicants.csv', index=False)
    #print(df.head())
else:
    print("No data fetched.")

df
```

### Using Endpoints - Applicants2Jobs

```python
# applicants2jobs endpoint
endpoint = 'applicants2jobs' 
df = fetch_data_from_endpoint(endpoint)

if df is not None:
    #print(df.head())
    applicants2jobs_df = df
    print("Columns in applicants:", applicants2jobs_df.columns)
    # Filter for hired and interviewed candidates
    hired_candidates = applicants2jobs_df[applicants2jobs_df['workflow_step_id'] == 'workflow_step_hired']
    interview_candidates = applicants2jobs_df[applicants2jobs_df['workflow_step_id'] == 'workflow_step_interview']

    # Identify hired/interviewed applicants by the workflow step ID
    hired_applicant_ids = hired_candidates['applicant_id'].unique()
    interview_applicant_ids = interview_candidates['applicant_id'].unique()
    
    # new column created in applicants_df that indicates whether the applicant is hired or interviewed
    applicants_df['hired_candidates'] = applicants_df['id'].apply(lambda x: 'Yes' 
                                                                  if x in hired_applicant_ids 
                                                                  else 'No')
    applicants_df['interview_candidates'] = applicants_df['id'].apply(lambda x: 'Yes' 
                                                                      if x in interview_applicant_ids 
                                                                      else 'No')
    applicants_df.to_csv('fetched_applicants2jobs.csv', index=False)

else:
    print("No data fetched.")
    
applicants_df
```

### Checking the Columns

```python
import pandas as pd

applicants_df = pd.read_csv('fetched_applicants.csv')
applicants2jobs_df = pd.read_csv('fetched_applicants2jobs.csv')
jobs_df = pd.read_csv('fetched_jobs.csv')

print("Columns in applicants DataFrame:", applicants_df.columns)
print("Columns in applicants2jobs DataFrame:", applicants2jobs_df.columns)
print("Columns in jobs DataFrame:", jobs_df.columns)
```

### Merging Applicants2Jobs and Jobs Info

```python
# Rename 'id' in jobs_df to 'job_id' to match the corresponding column in applicants2jobs_df
jobs_df.rename(columns={'id': 'job_id'}, inplace=True)

# merge
merged_df = applicants2jobs_df.merge(jobs_df, on='job_id', how='left')

# Merge with applicants_df to bring in applicant information
final_df = merged_df.merge(applicants_df, left_on='id', right_on='id', how='left')

# Save to a new CSV
final_df.to_csv('final_table.csv', index=False)
final_df
```

### Analysis of JazzHR data

```python
# Step 1: Count all applicants per job
applicant_counts = final_df['job_id_x'].value_counts().reset_index()
applicant_counts.columns = ['job_id_x', 'applicant_count']

# Step 2: Filter for hired candidates and count them per job
hired_applicants = final_df[final_df['hired_candidates'] == 'Yes']
hired_counts = hired_applicants['job_id_x'].value_counts().reset_index()
hired_counts.columns = ['job_id_x', 'hired_count']

# Step 3: Filter for interviewed candidates and count them per job
interviewed_applicants = final_df[final_df['interview_candidates'] == 'Yes']
interviewed_counts = interviewed_applicants['job_id_x'].value_counts().reset_index()
interviewed_counts.columns = ['job_id_x', 'interviewed_count']

# Step 4: Merge the hired and interviewed counts back into the applicant_counts DataFrame
final_counts = pd.merge(applicant_counts, hired_counts, on='job_id_x', how='left')
final_counts = pd.merge(final_counts, interviewed_counts, on='job_id_x', how='left')

# Step 5: Fill NaN values with 0 for 'hired_count' and 'interviewed_count' columns
final_counts['hired_count'] = final_counts['hired_count'].fillna(0).astype(int)
final_counts['interviewed_count'] = final_counts['interviewed_count'].fillna(0).astype(int)

# Print the final DataFrame to check
print(final_counts)
```
